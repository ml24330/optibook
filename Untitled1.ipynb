```python
import pandas as pd

df = pd.read_csv('data.csv', index_col=0)
df.reset_index(inplace=True, drop=True)
```

```python
import matplotlib.pyplot as plt

normalised = pd.DataFrame()

for column in df:
    normalised[column] = df[column]/df[column].iloc[0]
    plt.plot(normalised[column], label=column)

df.corr()
```

```python
import seaborn as sns

sns.pairplot(df)
plt.savefig("img1")
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

diff_norm = normalised["TOTAL"] - normalised["UNILEVER"]
diff_abs = df["TOTAL"] - df["UNILEVER"]
# plt.plot(normalised["TOTAL"])
# plt.plot(normalised["UNILEVER"])
model = ARIMA(diff_norm.head(5000), order=(2,0,0)).fit()
plt.plot(model.predict(start=0,end=5000))
plt.plot(diff_norm.head(5000))
```

```python
def rolling_forecast(df, test_size, order):
    forecast = list()

    for i in range(0,test_size):
        new_size = test_size-i
        train, test = df[:len(df)-new_size], df[len(df)-new_size:]
        model = ARIMA(train, order=order).fit()
        forecast.append(model.forecast().iloc[-1])

    forecast = df[:len(df)-test_size].append(pd.Series(forecast))
    forecast = forecast.reset_index()[0]
    
    return forecast

forecast = rolling_forecast(diff_norm.head(400), 30, order=(1,0,0))
plt.plot(forecast.iloc[300:400], label="Forecast")
plt.plot(diff_norm.iloc[300:400], label="Actual")
plt.legend()
```
